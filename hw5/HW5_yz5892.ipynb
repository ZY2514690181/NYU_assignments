```python
import os

from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
```

```python
# Training file

f1 = open('WSJ_02-21.pos-chunk')
lines = f1.read()
f1.close()

lines = lines.split('\n')

# Write to file

f2 = open("training.feature", "w")

words = []
POSs = []
tags = []
count = 0
written = ''
for line in lines:
    line = line.split('\t')
    if len(line) > 1:
        # This is not a blank line
        words.append(line[0])
        POSs.append(line[1])
        tags.append(line[2])
        count += 1
    else:
        # This is a blank line
        for i in range(count):
            written = written + words[i] + '\t'
            written = written + 'POS=' + POSs[i] + '\t'
            if i >= 3:
                written = written + 'pre3_word=' + words[i-3] + '\t'
                written = written + 'pre3_POS=' + POSs[i-3] + '\t'
            if i >= 2:
                written = written + 'pre2_word=' + words[i-2] + '\t'
                written = written + 'pre2_POS=' + POSs[i-2] + '\t'
            if i >= 1:
                written = written + 'pre1_word=' + words[i-1] + '\t'
                written = written + 'pre1_POS=' + POSs[i-1] + '\t'
            if i <= count - 2:
                written = written + 'next1_word=' + words[i+1] + '\t'
                written = written + 'next1_POS=' + POSs[i+1] + '\t'
            if i <= count - 3:
                written = written + 'next2_word=' + words[i+2] + '\t'
                written = written + 'next2_POS=' + POSs[i+2] + '\t'
            if i <= count - 4:
                written = written + 'next3_word=' + words[i+3] + '\t'
                written = written + 'next3_POS=' + POSs[i+3] + '\t'
            written = written + tags[i] + '\n'
        written += '\n'
        f2.write(written)
        words = []
        POSs = []
        tags = []
        count = 0
        written = ''
# end for
f2.close()
```
