```python
f = open("dollar_output.txt", "w")



some_num1 = '(one|One|two|Two|three|Three|four|Four|five|Five|six|Six|seven|Seven|eight|Eight|nine|Nine|ten|Ten|eleven|Eleven|twelve|Twelve|thirteen|Thirteen|fourteen|Fourteen|fifteen|Fifteen|sixteen|Sixteen|seventeen|Seventeen|eighteen|Eighteen|nineteen|Nineteen|twenty|Twenty|thirty|Thirty|forty|Forty|fifty|Fifty|sixty|Sixty|seventy|Seventy|eighty|Eighty|ninety|Ninety|half|quarter|half of|quarter of|a|A)'
some_num2 = '(one|One|two|Two|three|Three|four|Four|five|Five|six|Six|seven|Seven|eight|Eight|nine|Nine|ten|Ten|eleven|Eleven|twelve|Twelve|thirteen|Thirteen|fourteen|Fourteen|fifteen|Fifteen|sixteen|Sixteen|seventeen|Seventeen|eighteen|Eighteen|nineteen|Nineteen|twenty|Twenty|thirty|Thirty|forty|Forty|fifty|Fifty|sixty|Sixty|seventy|Seventy|eighty|Eighty|ninety|Ninety|half|quarter|half of|quarter of|a|A|hundred|Hundred|thousand|Thousand|million|Million|billion|Billion)'



# $ Arabic numerals
# This considers $ Arabic numerals and $ followed by letter numerals: output1 = re.finditer('(\$\s*([0-9,]+\.?[0-9,]+\sbillion)?([0-9,]+\.?[0-9,]+\smillion)?([0-9,]+\.?[0-9,]+\sthousand)?([0-9,]+\.?[0-9,]+\shundred)?) | (\$\s*('+some_num+'\sbillion)?('+some_num+'\smillion)?('+some_num+'\sthousand)?('+some_num+'\shundred)?)', lines)
output1 = re.finditer('\$\s*([0-9,]+(\.[0-9,]+)?\sbillion)?([0-9,]+(\.[0-9,]+)?\smillion)?([0-9,]+(\.[0-9,]+)?\sthousand)?([0-9,]+(\.[0-9,]+)?\shundred)?([0-9,]+(\.[0-9,]+)?)?', lines)

for item in output1:
    print(item.group(0))
    f.write(item.group(0))
    f.write('\n')


# Arabic numerals dollar/cent

output2 = re.finditer('[0-9]([0-9,]*(\.[0-9,]*)?\sbillion)?([0-9,]*(\.[0-9,]*)?\smillion)?([0-9,]*(\.[0-9,]*)?\sthousand)?([0-9,]*(\.[0-9,]*)?\shundred)?([0-9,]*(\.[0-9,]*)?)?\s*dollar(s)?\s*([0-9]*cent(s)?)?', lines)
for item in output2:
    print(item.group(0))
    f.write(item.group(0))
    f.write('\n')


# word dollar

#output3 = re.finditer('('+some_num+'\sbillion\s)?('+some_num+'\smillion\s)?('+some_num+'\sthousand\s)?('+some_num+'\shundred\s)?('+some_num+')?\s*dollar(s)?', lines)
output3 = re.finditer('('+some_num1+')+\s*('+some_num2+')*\s*dollar(s)?', lines)
for item in output3:
    print(item.group(0))
    f.write(item.group(0))
    f.write('\n')

# word cent
#output4 = re.finditer('('+some_num+'\sbillion\s)?('+some_num+'\smillion\s)?('+some_num+'\sthousand\s)?('+some_num+'\shundred\s)?('+some_num+')?\s*dollar(s)?', lines)
output4 = re.finditer('('+some_num1+')+\s*('+some_num2+')*\s+cent(s)?', lines)
for item in output4:
    print(item.group(0))
    f.write(item.group(0))
    f.write('\n')

f.close()
```
